# Decorator

- Decorator là một hàm nhận đầu vào là một fuction và trả về một function
- mọi thứ trong python đều là object
- Thay đổi hành vi của một hàm số mà không cần sửa code của hàm số đó.
  

````python
def add10(func):
    def inner(*args):
        return func(*args) + 10
    return inner


@add10
@add10
@add10
def add(a, b):
    return a + b

@add10
@add10
@add10
@add10
def tich_a_vs_b(a, b):
    return a * b

print(tich_a_vs_b(3, 3))
print(add(3, 3))

````
